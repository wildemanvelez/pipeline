```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
names= ['buying',
    'maint',
'doors',
'persons',
'lug_boot',
'safety ',]
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data', names=names)
```

```python
data.head()
```

```python
x=data.iloc[:,:-1]
```

```python
y=data.iloc[:,-1]
```

```python
x
```

```python
x = pd.get_dummies(x)
```

```python
x
```

```python
y = pd.get_dummies(y)
```

```python
y
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
model=MLPRegressor(hidden_layer_sizes=(80,80,20),max_iter=5000000)
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict(x)
```

```python
plt.plot(model.predict(x), 'ob')
plt.plot(y,'ob')
```
